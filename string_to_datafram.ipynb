```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Load the dataset

```

```python
lst= "i am gotta be golden i am gotta be golden i am gotta be golden GOLDEN" 
dict1 = {}
for i in lst:
    if 'A'<=i<='Z':
       print(i)
       i = chr(ord(i)+32)
    if 'a'<=i <='z':
        if i in dict1:
            dict1[i] +=1
        else:
            dict1[i] =1
print(dict1)
```

```python
# Convert dictionary items to a list of tuples

items = list(dict1.items())

# Define column names for the DataFrame
column_names =['letters','count']

#create a DataFrame from the list of tuples
df_letters= pd.DataFrame(items,columns=column_names)
df_letters.sort_values(by='letters', inplace=True)
print(df_letters)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
bars=plt.bar(df_letters['letters'], df_letters['count'], color='skyblue')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval), ha='center', va='bottom')
plt.title('letter count')
plt.xlabel('letters')
plt.ylabel ('count')
plt.show()
```
