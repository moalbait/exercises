```python
import numpy as np

# x = input (e.g. temperature)
x = np.array([1,2,3,4,5])

# y = output (e.g. pollution)
y = np.array([2,4,5,4,5])

# Calculate the mean of x and y

mean_x= np.mean(x)
mean_y = np.mean(y)

# Calculate the slope (m) and intercept (b) for the line of best fit
m_upper = np.sum((x-mean_x)*((y-mean_y)))
m__lower = np.sum((x-mean_x)**2)
m = m_upper/m__lower

# Calculate the intercept (b)
b= mean_y-m*mean_x
print(f"slope: {m}, intercept: {b}")
```

```python
#b=yˉ​−mxˉ
y_pred= m*x+b
print(f"Predicted y values: {y_pred}")
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,y_pred, color='red')
plt.title("Linear Regression")
plt.xlabel("x")
plt.ylabel("y_pred")
plt.show()
```

```python
#mean squared Error
mse = np.mean((y-y_pred)**2)
print(f"Mean Squared Error: {mse}")
```
